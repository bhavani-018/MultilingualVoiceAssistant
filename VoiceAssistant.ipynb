```python
import speech_recognition as sr 
from gtts import gTTS
import playsound 
import translators as ts
import os 
import pyaudio
```

```python
#  convert speech to text - English
def dobby_listen_en():
    # create recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source:
        # use the listen function so the recognizer can catch the source (our microphone)
        audio = r.listen(source)
        text = r.recognize_google(audio, language='en')

    text = text.lower()
    return text
```

```python
#  convert speech to text - Tamil
def dobby_listen_ta():
    # create recognizer
    r = sr.Recognizer()
    
    with sr.Microphone() as source:
        # use the listen function so the recognizer can catch the source (our microphone)
        audio = r.listen(source)
        text = r.recognize_google(audio, language='ta')
    
    text = text.lower()
    return text
```

```python
a = dobby_listen_ta()
print(a)
```

```python
# convert text to speech - English
def dobby_talk_en(text):
    # create audio
    file_name ='audio_data.mp3'
    # convert text to speech
    tts = gTTS(text = text, lang='en')
    # save file
    tts.save(file_name)
    # play file
    playsound.playsound(file_name)
    # remove file
    os.remove(file_name)

```

```python
dobby_talk_en('hii how are you? nice to meet you')
```

```python
# convert text to speech - Tamil
def dobby_talk_ta(text):
    # create audio
    file_name ='audio_data.mp3'
    # convert text to speech
    tts = gTTS(text = text, lang='ta')
    # save file
    tts.save(file_name)
    # play file
    playsound.playsound(file_name)
    # remove file
    os.remove(file_name)
```

```python
dobby_talk_ta('வணக்கம் எப்படி இருக்கீங்க நலமா? உங்களை சந்தித்ததில் மகிழ்ச்சி')
```

```python
# convert text to speech - Hindi
def dobby_talk_hi(text):
    # create audio
    file_name ='audio_data.mp3'
    # convert text to speech
    tts = gTTS(text = text, lang='hi')
    # save file
    tts.save(file_name)
    # play file
    playsound.playsound(file_name)
    # remove file
    os.remove(file_name)
```

```python
# Namaste, aap kaise hain?” → “Hello, how are you?”
dobby_talk_hi('Namaste, aap kaise hain?')
```

```python
def translate_en_hi(text):
    """English → Hindi"""
    translated = ts.translate_text(text, from_language='en', to_language='hi')
    dobby_talk_hi(translated)

def translate_en_ta(text):
    """English → Tamil"""
    translated = ts.translate_text(text, from_language='en', to_language='ta')
    dobby_talk_ta(translated)

def translate_ta_hi(text):
    """Tamil → English"""
    translated = ts.translate_text(text, from_language='ta', to_language='hi')
    dobby_talk_ta(translated)


def dobby_reply(text):
  
    if 'translate' in text:
        while True:
            dobby_talk_en('I will choose the right translator for you. Let me know the source language and the target language.')
            source_target_lang = dobby_listen_en()
            print(source_target_lang)


            if 'english to hindi' in source_target_lang:

                dobby_talk_en('Got it, you need a translator from English to Hindi. What can I translate for you?')
                while True:
                    text_to_translate = dobby_listen_en()
                    print(text_to_translate)
        
                if text_to_translate != 'switch translator':
                    translate_en_hi(text_to_translate)
                else: 
                    break

            elif 'english to tamil' in source_target_lang:
                dobby_talk_en('Got it, you need a translator from English to Tamil. What can I translate for you?')
                while True:
                    text_to_translate = dobby_listen_en()
                    print(text_to_translate)
        
                if text_to_translate != 'switch translator':
                    translate_en_ta(text_to_translate)
                else: 
                    break

            elif 'tamil to hindi' in source_target_lang:
                dobby_talk_ta('Got it, you need a translator from Tamil to Hindi. What can I translate for you?')
                while True:
                    text_to_translate = dobby_listen_ta()
                    print(text_to_translate)
        
                if text_to_translate != 'switch translator':
                    translate_ta_hi(text_to_translate)
                else: 
                    break

            elif 'turn off translator':
                dobby_talk_en('It was a pleasure to do the translation job. Wish you a great day!')
                break
            
            else:
                dobby_talk_en('Sorry, I did not get what you said')
      
    
    else:
        dobby_talk_en('Sorry, I can not support you with that.')

            


```

```python
def dobby_run():
    dobby_talk_en('Hi there. Nice to meet you. I am Dobby, your personal translator. What is your name?')
    listen_name = dobby_listen_en()
    dobby_talk_en('Hi ' + listen_name + ' how can I help you?')
    
    listen_assistant = dobby_listen_en()
    print(listen_assistant)
    dobby_reply(listen_assistant)
    
dobby_run()  
```
